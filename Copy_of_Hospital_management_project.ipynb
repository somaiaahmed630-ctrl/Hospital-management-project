```python
ds.dtypes
```

```python
ds['START'] = pd.to_datetime(ds['START'], errors='coerce')
ds['STOP'] = pd.to_datetime(ds['STOP'], errors='coerce')
```

```python
ds.dtypes
```

```python
total_cost = ds['BASE_ENCOUNTER_COST'].sum()
print("Total BASE_ENCOUNTER_COST:", total_cost)
```

```python
total_claim_cost = ds['TOTAL_CLAIM_COST'].sum()
print("Total_CLAIM_COST:", total_claim_cost)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=ds['BASE_ENCOUNTER_COST'])
plt.title('Outliers in Encounter Cost')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(ds['BASE_ENCOUNTER_COST'], bins=30, kde=True)
plt.title('Distribution of Base Encounter Cost')
plt.xlabel('Cost')
plt.ylabel('Frequency')
plt.show()
```
